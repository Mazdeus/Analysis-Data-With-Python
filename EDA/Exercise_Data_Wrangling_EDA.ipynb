# CLEANING DATA

1. MEMBERSIHKAN DATA CUSTOMERS_DF

```python
# Menghilangkan duplikat data
customers_df.drop_duplicates(inplace=True)
```

```python
print("jumlah duplicate : ", customers_df.duplicated().sum())
```

MENANGANI MISSING VALUE

```python
# Data yang mengandung missing value
customers_df[customers_df.gender.isna()]
```

Banyak informasi penting, sayang jika dibuang, jadi gunakan metode imputation. Gunakan nilai dominan sebagai pengganti missing value.

```python
# Cari nilai dominan
customers_df.gender.value_counts()
```

```python
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
# Cek missing value
customers_df.isna().sum()
```

MENGANGANI INACCURATE VALUE

```python
# Lihat data max
customers_df[customers_df.age == customers_df.age.max()]
```

Kita bisa berasumsi bahwa inaccurate value tersebut terjadi karna human error sehingga kelebihan memasukkan nilai 0. Ganti jadi 70.

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# Cek
customers_df[customers_df.age == customers_df.age.max()]
```

Ternyata masih ada invalid value lain

```python
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
customers_df.describe()
```

Eksplorasi data customers_df

```python
# Eksplorasi Data customers_df
customers_df.describe(include="all")
```

```python
# Demografi pelanggan berdasarkan jenis kelamin
customers_df.groupby(by="gender").agg({
    "customer_id": "count",
    "age": ["max","min","mean","std"]
})
```

```python
# Persebaran jumlah pelanggan berdasarkan kota (city) dan negara bagian (state)
customers_df.groupby(by="city").customer_id.nunique().sort_values(ascending=False)
```

```python
customers_df.groupby(by="state").customer_id.nunique().sort_values(ascending=False)
```

Eksplorasi data orders_df

```python
# Membuat kolom baru untuk memuat waktu pengiriman tiap order
# Selisih antara delivery_date dan order_date
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df
```

```python
orders_df.describe(include="all")
```

Eksplorasi data orders_df dan customers_df

```python
#  Membuat kolom baru bernama "status" pada data customers_df untuk mengidentifikasi pelanggan yang belum pernah order.
customer_id_in_orders_df = orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Nonactive")
customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

```python
# Merge  data orders_df dan customers_df yaitu left
orders_customers_df = pd.merge(
    left = orders_df,
    right = customers_df,
    how = "left",
    left_on = "customer_id",
    right_on = "customer_id"
)
orders_customers_df
```

```python
# Jumlah Order berdasarkan kota
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
# Jumlah order berdasarkan state
orders_customers_df.groupby(by = "state").order_id.nunique().sort_values(ascending=False)
```

```python
# Jumlah order berdasarkan gender
orders_customers_df.groupby(by = "gender").order_id.nunique().sort_values(ascending=False)
```

```python
# Jumlah order berdasarkan kelompok usia
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x : "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
orders_customers_df.groupby(by = "age_group").order_id.nunique().sort_values(ascending=False)
```

Eksplorasi data product_df dan  sales_df

```python
product_df.describe(include="all")
```

```python
sales_df.describe(include="all")
```

```python
product_df.sort_values(by="price", ascending=False)
```

```python
# Pivot table untuk mencari informasi terkait produk berdasarkan tipe
product_df.groupby(by="product_type").agg({
    "product_id":"nunique",
    "quantity":"sum",
    "price":["min","max"]
})
```

```python
# Pivot table untuk mencari informasi terkait produk berdasarkan nama produknya.
product_df.groupby(by="product_name").agg({
    "product_id":"nunique",
    "quantity":"sum",
    "price":["min","max"]
})
```

```python
# Untuk mencaari produk yang paling laris, merge tabel product_df dan sales_df
sales_product_df = pd.merge(
    left = sales_df,
    right = product_df,
    how = "left",
    left_on = "product_id",
    right_on = "product_id"
)
sales_product_df.head()
```

```python
# Melihat informasi penjualan product berdasarkan tipenya.
sales_product_df.groupby(by = "product_type").agg({
    "sales_id" : "nunique",
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
# Melihat informasi penjualan product berdasarkan nama product
sales_product_df.groupby(by = "product_name").agg({
    "sales_id" : "nunique",
    "quantity_x" : "sum",
    "total_price" : "sum"
}).sort_values(by = "total_price", ascending=False)
```

Eksplorasi data all_df

```python
# Penggabungan
all_df = pd.merge(
    left = sales_product_df,
    right = orders_customers_df,
    how = "left",
    left_on = "order_id",
    right_on = "order_id"
)
all_df.head()
```

```python
# Melihat preferensi pembelian berdasarkan state pelanggan dan tipe produk
all_df.groupby(by = ["state", "product_type"]).agg({
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
# Selera tipe produk pelanggan berdasarkan gender
all_df.groupby(by = ["gender", "product_type"]).agg({
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
# Selera tipe produk pelanggan berdasarkan usia
all_df.groupby(by = ["age_group", "product_type"]).agg({
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```
