# **C. DATA VISUALIZATION**

1. Bagaimana Performa Penjualan dan Revenue Perusahaan dalam Beberapa Bulan Terakhir?

```python
# Ubah frekuensi dari data yang awalnya harian menjadi bulanan
monthly_orders_df = all_df.resample(rule = "M", on = 'order_date').agg({
    "order_id" : "nunique",
    "total_price" : "sum"
})

monthly_orders_df.index = monthly_orders_df.index.strftime("%Y-%m")
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id" : "order_count",
    "total_price" : "revenue"
}, inplace = True)

monthly_orders_df.head()
```

```python
# Visualisasi line chart untuk jumlah order
plt.figure(figsize=(10,5))
plt.plot(monthly_orders_df["order_date"], monthly_orders_df["order_count"], marker = 'o', linewidth = 2, color = "#72BCD4")
plt.title("Number of Orders per Month (2021)", loc = "center", fontsize = 20)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.show()
```

```python
# Visualisasi line chart untuk revenue
plt.figure(figsize=(10,5))
plt.plot(
    monthly_orders_df["order_date"],
    monthly_orders_df["revenue"],
    marker = "o",
    linewidth = 2,
    color="#72BCD4"
)

plt.title("Total revenue per month (2021)", loc = "center", fontsize = 20)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.show()
```

Untuk 3 pertanyaa terakhir pakai teknik RFM Analysis.
RFM Analysis digunakan untuk melakukan segmentasi pelanggan berdasarkan tiga paramter yaitu :
- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
- Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
- Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
rfm_df = all_df.groupby(by = "customer_id", as_index = False).agg({
    "order_date" : "max", # Mengambil tanggal order terakhir
    "order_id" : "nunique", # Menghitung jumlah order
    "total_price" : "sum" # menghitung jumlah revenue yang dihasilkan
})

# Mengganti nama kolom hasil agregasi agar sesuai dengan terminologi RFM.
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# Menghitung kapan terakhir pelanggan melakukan transaksi (hari). Mengubah tipe data kolom 'max_order_timestamp' dari datetime menjadi date (menghilangkan informasi waktu).
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date

# Mencari tanggal transaksi paling baru dari keseluruhan data sebagai acuan perhitungan recency.
recent_date = orders_df["order_date"].dt.date.max()

# Menghitung recency (keterbaruan) dengan mengurangkan tanggal transaksi terakhir setiap pelanggan dari 'recent_date'.
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis = 1, inplace = True)
rfm_df.head()
```

```python
# Mengidentifikais best customer
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (30,6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y = "recency", x = "customer_id", data = rfm_df.sort_values(by = "recency", ascending = True).head(5), palette = colors, ax = ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Customer by Recency", loc = "center", fontsize = 18)
ax[0].tick_params(axis = "x", labelsize = 15)

sns.barplot(y = "frequency", x = "customer_id", data = rfm_df.sort_values(by = "frequency", ascending = False).head(5), palette = colors, ax = ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("Best Customer by Frequency", loc = "center", fontsize = 18)
ax[1].tick_params(axis = "x", labelsize = 15)

sns.barplot(y = "monetary", x = "customer_id", data = rfm_df.sort_values(by = "monetary", ascending = False).head(5), palette = colors, ax = ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("Best Customer by Monetary", loc = "center", fontsize = 18)
ax[2].tick_params(axis = "x", labelsize = 15)

plt.suptitle("Best Customer by RFM Parameters (customer_id)", fontsize = 30)
plt.show()
```
