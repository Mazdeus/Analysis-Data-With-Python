# CLEANING DATA

1. MEMBERSIHKAN DATA CUSTOMERS_DF

```python
# Menghilangkan duplikat data
customers_df.drop_duplicates(inplace=True)
```

```python
print("jumlah duplicate : ", customers_df.duplicated().sum())
```

MENANGANI MISSING VALUE

```python
# Data yang mengandung missing value
customers_df[customers_df.gender.isna()]
```

Banyak informasi penting, sayang jika dibuang, jadi gunakan metode imputation. Gunakan nilai dominan sebagai pengganti missing value.

```python
# Cari nilai dominan
customers_df.gender.value_counts()
```

```python
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
# Cek missing value
customers_df.isna().sum()
```

MENGANGANI INACCURATE VALUE

```python
# Lihat data max
customers_df[customers_df.age == customers_df.age.max()]
```

Kita bisa berasumsi bahwa inaccurate value tersebut terjadi karna human error sehingga kelebihan memasukkan nilai 0. Ganti jadi 70.

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# Cek
customers_df[customers_df.age == customers_df.age.max()]
```

Ternyata masih ada invalid value lain

```python
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
customers_df.describe()
```

2. MEMBERSIHKAN DATA ORDERS_DF

```python
# Mengganti tipe data pada kolom order_date dan delivery_date menjadi datetime.
datetime_columns = ["order_date", "delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
# Cek kembali
orders_df.info()
```
